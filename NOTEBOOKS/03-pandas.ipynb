- tailnum: Plane tail number
- flight: Flight number
- origin,dest: Origin and destination.

- distance: Distance flown

## Combinando datasets

Existem duas fun√ß√µes principais, o *join()* e o *merge()*.

A principal diferen√ßa entre *join* e *merge* √© que o *join()* √© usado para combinar dois DataFrames pelo √≠ndice, mas n√£o em colunas, enquanto *merge()* √© usado principalmente para especificar as colunas que voc√™ deseja unir.

https://pandas.pydata.org/docs/reference/api/pandas.merge.html

```python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=1)
```
